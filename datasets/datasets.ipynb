```python
import arxiv
import pprint
import semanticscholar as sch
```

## arXiv

```python
papers = arxiv.query(
    query="cat:cs.LG",
    max_results=1,
    iterative=True
)

paper_keys = set()

for paper in papers():
    pprint.pprint(paper)
    for key in paper.keys():
        if key not in paper_keys:
            paper_keys.add(key)
            
print(paper_keys)
print(len(paper_keys))
```
